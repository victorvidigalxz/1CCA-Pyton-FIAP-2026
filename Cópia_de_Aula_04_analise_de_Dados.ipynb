```python
# Estatisticas descritivas das colunas ---->
# Exemplo: Voltage (Tensao, em Volts) e Global_intensity (Correntes, em Aperes)

# Media aritimetica ------> mean
media_tensao = tabela ["Voltage"].mean()
print(f"A Media da tensao sera de: {media_tensao:.2f}V")
# Mediana -------> median
mediana_tensao = tabela ["Voltage"].median()
print(f"A Mediana da tensao sera de: {mediana_tensao:.2f}V")
#Desvio Padrao --------> standart Deviatin std
desvio_padrao = tabela ["Voltage"].std()
print(f"O desvio padrao sera de: {desvio_padrao:.2f}V")

```

```python
# Estatisticas descritivas das colunas ---->
# Exemplo: Voltage (Tensao, em Volts) e Global_intensity (Correntes, em Aperes)

# Media aritimetica ------> mean
media_tensao = tabela ["Global_intensity"].mean()
print(f"A Media da tensao sera de: {media_tensao:.2f}A")
# Mediana -------> median
mediana_tensao = tabela ["Global_intensity"].median()
print(f"A Mediana da tensao sera de: {mediana_tensao:.2f}A")
#Desvio Padrao --------> standart Deviatin std
desvio_padrao = tabela ["Global_intensity"].std()
print(f"O desvio padrao sera de: {desvio_padrao:.2f}A")
```
